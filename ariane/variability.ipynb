```python
# booleans to separate into the different sections
def boo(d,section):
    high = 33.5 # salinity division between CUC and south shelf
    low = 32 # salinity division between south shelf and Columbia

    if section == 'north':
        boolean = d.final_section == 4
    elif section == 'offshore':
        boolean = d.final_section == 3
    elif section == 'south':
        boolean = (d.final_section ==2) & (d.final_salt < high) & (d.final_salt >= low)
    elif section == 'cuc':
        boolean = (d.final_section ==2) & (d.final_salt >= high)
    elif section == 'brackish':
        boolean = (d.final_section ==2) & (d.final_salt < low)
    elif section == 'meander':
        boolean = (d.final_section ==0) & ((d.init_t-d.final_t)>24)
    else:
        print('section options are north,offshore,south,cuc,brackish,meander')
    
    return boolean
```

```python
def running_mean(x, N):

    # Compute the cumulative sum while ignoring NaNs.
    cumsum = np.nancumsum(np.insert(x, 0, 0)) 
    # Create a boolean array where True represents non-NaN elements.
    non_nan = np.insert(~np.isnan(x), 0, 0)
    # Compute the cumulative count of non-NaN elements.
    count = np.cumsum(non_nan)
    
    # Calculate the total sum for each window.
    total_sum = cumsum[N:] - cumsum[:-N]
    # Calculate the count of non-NaN elements in each window.
    valid_count = count[N:] - count[:-N]
    
    # Use np.divide to safely divide, turning 0/0 into NaN instead of raising an error.
    return np.divide(total_sum, valid_count, out=np.zeros_like(total_sum), where=valid_count!=0)
```
